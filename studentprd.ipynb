#Drop unnamed colomn


```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

#data transformation
#change weekly styudy hours column


```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

#gender distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

#gender distribution

```python
plt.figure(figsize=(4,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above data females count is more than male
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,5))
sns.heatmap(gb,annot=True)
plt.title("Relation between parent's Education and student's score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parent have good impact on your course
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,5))
sns.heatmap(gb1,annot=True)
plt.title("Relation between parent's Marital status and student's score")

plt.show()
```

```python
#from the above chart we have Conclueded that 
#marital status of parental has no impact or negligable impact on the student score
```

#Distributiob of Ethnic group

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()
l=["groupA","groupB","groupC","groupD","groupE"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

```python
ax=sns.countplot(data=df,x='EthnicGroup',hue='EthnicGroup',palette='Set2',dodge=False)
ax.bar_label(ax.containers[0])
```

```python
#hue tells  seaborn to color each group differently
#dodge=false aviod bars splitting into two (keep one bar group)
```
